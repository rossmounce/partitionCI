Naming scheme: be explicit! 
Start with whole TNT files as .tntwhole & partition data files as either .tntcran & .tntpost OR .nexcran & .nexpost

**[In a _for_ loop] Do whole dataset searches in TNT, export files&trees as nexus**

```python
!for i in *.tntwhole; do tnt mxram 600, p $i , hold 500000, xinact, xmult=level10, bb, export *${i%.*}.nexwhole, quit, ; done
```

**[In a _for_ loop] Cut-off whole dataset MPT blocks**

```python
!for i in *.nexwhole; do sed -n '/begin trees ;/,/end ;/p' $i > ${i%.*}.tre ; done
```

**[In a _for_ loop] Check number of whole data analysis MPTs**

```python
!for i in *.tre ; do grep -c -H 'tree tnt' $i ; done
```

**[In _for_ loops] If partition data file exists as a TNT file, convert to Nexus**

```python
!for i in *.tntcran; do tnt mxram 600, p $i , xinact , export ${i%.*}.nexcran, quit, ; done
```

```python
!for i in *.tntpost; do tnt mxram 600, p $i , xinact , export ${i%.*}.nexpost, quit, ; done
```

**[In _for_ loops] Add tree block from WHOLE analysis to NEXUS partition data files**

```python
!for i in *.nexcran; do cat $i ${i%.*}.tre > ${i%.*}.nexcrantree ; done
```

```python
!for i in *.nexpost; do cat $i ${i%.*}.tre > ${i%.*}.nexposttree ; done
```

```python
!for i in *.nexcrantree; do sed -i -e "\$abegin paup; exclude uninf; pscores /ci=yes ri=yes single=all scorefile=${i%.*}.cranlog ; end;" $i ; done
```

```python
!for i in *.nexposttree; do sed -i -e "\$abegin paup; exclude uninf; pscores /ci=yes ri=yes single=all scorefile=${i%.*}.postlog ; end;" $i ; done
```
